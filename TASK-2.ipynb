# **Importing Libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb
from sklearn.datasets import load_iris

%matplotlib inline
```

# **Load the Dataset**

```python
#loaidng the iris dataset
iris = load_iris() 
iris.data
```

# **Plot the data as Pair Plot**

```python
sb.set(style = 'whitegrid')
plt.figure(figsize = (20, 20))
sb.pairplot(df, plot_kws = {'color' : 'blue'});
```

# **Plotting Heat Map of Correlated Data**

```python
sb.heatmap(df.corr(),annot=True)
```

```python
X=df.iloc[:,:].values #creating array of dependent variables
```

# **Using K-means Clustering Algorithm**

```python
from sklearn.cluster import KMeans
wcss = [] # Within cluster sum of squares
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', 
                    max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow method')
plt.xlabel('No. of clusters')
plt.ylabel('WCSS') 
plt.show()
```

# **Training the Model**

```python
# Applying kmeans to the dataset / Creating the kmeans classifier
kmeans = KMeans(n_clusters = 3, init = 'k-means++',
                max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(X)
print('training completed')
```

# **Visualising the clusters & Plot Centroids**

```python
#Visualising the clusters
plt.scatter(X[:,0], X[:,1])
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')
plt.show()
```

```python
# Visualising the clusters - On the first two columns
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'orange', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], 
            s = 100, c = 'purple', label = 'Centroids')

plt.legend()
```
